# PWC estimator algorithm

## Python implementation examples

```python
# import required libraries 
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
# testDefinitions is required for generating the sample of example 2).
import testDefinitions as tf
# import PWC dstribution class definition script
import pwc as pwc
```

## 1)

```python
# PWC estimator algorithm run on data generated from a distrution composed
# by bivariate uniform distributions (a discrete mixture of uniforms with equal weights):
# 1) uniform on [-1, 1]x[-1, 1]
# 2) uniform on [6, 8]x[10, 33]
# 3) uniform on [1, 2]x[1, 2]
# 4) uniform on [2, 2.5]x[10, 24]

np.random.seed(22)
u1 = np.random.uniform(-1, 1, size=(300, 2))
u2 = np.concatenate((np.random.uniform(6, 8, size=(300, 1)),
                    np.random.uniform(10, 33, size=(300, 1))), axis=1)
u3 = np.random.uniform(1, 2, size=(300, 2))
u4 = np.concatenate((np.random.uniform(2, 2.5, size=(300, 1)),
                    np.random.uniform(10, 24, size=(300, 1))), axis=1)
xs = np.concatenate((u1, u2, u3, u4), axis=0)
```

```python
# create PWC distribution class object and run method to build PWC estimator
pwc1 = pwc.PWCDistribution(xs)
pwc1.PWCEstimator(verbose=False)
pwc1.plot(showdata=True);
```

## 2)

```python
# PWC estimator algorithm run on the sample used in Section 7 of the paper.
# For more detail see the above mentioned section of the paper and refer to the original work:
# O’Brien, Kashinath, Cavanaugh, Collins, and O’Brien (2016). A fast and objective multidimensional kernel density estimation method: fastkde. Computational Statistics & Data Analysis, 148-160.

np.random.seed(12)
test_mix = tf.testMixtureModel()
mix_samples = test_mix.sampleFromDistribution(int(1e6)).T
```

```python
# create PWC distribution class object and run method to build PWC estimator
pwc2 = pwc.PWCDistribution(mix_samples)
pwc2.PWCEstimator(verbose=False)
pwc2.plot(showdata=False, xlim=[-39, 39], ylim=[-39, 39]);
```
